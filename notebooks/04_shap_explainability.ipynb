### Objective
Use **SHAP (SHapley Additive exPlanations)** to interpret the
final trained credit-risk model.


```python
# Importing Modules
import pandas as pd
import numpy as np
import joblib
import matplotlib.pyplot as plt
import shap
```

```python
# Loading Dataset
model=joblib.load("../models/best_model.pkl")

X_Train,X_Test,Y_Train,Y_Test=joblib.load('../data/processed_data.pkl')
print("X Train Shape: ",X_Train.shape)
print("X Test Shape: ",X_Test.shape)
```

```python
# SHAP Explainer
explainer=shap.Explainer(model,X_Train)
shap_value=explainer(X_Test)
```

```python
shap.summary_plot(shap_value,X_Test)
```

```python
shap.plots.bar(shap_value)
```

```python
i = 0
shap.plots.waterfall(shap_value[i])
```

```python
shap.initjs()
shap.plots.force(shap_value[i])
```

### Key Explainability Insights

- Repayment delay features are the **strongest predictors** of default risk.
- High credit utilization increases probability of default.
- Consistent and timely payments reduce predicted risk.
- The model provides **transparent reasoning** for each lending decision,
  which is critical in financial applications.
