## Credit Risk Prediction - Exploratory Data Analysis (EDA)

### Objective:
This notebook explores the **Taiwan Credit Card Default dataset** to understand
data structure, missing values, class imbalance, and key factors affecting
**credit default risk**.

The insights from this analysis will support **feature engineering and machine
learning modeling** in later stages.

```python
# Importing Modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plot Setting
sns.set(style="whitegrid")
plt.rcParams['figure.figsize']=(8,5)
```

```python
# Loading Dataset
file_path="../data/credit_card_default_dataset.csv"

df=pd.read_csv(file_path,header=1)
df.head()
```

```python
# Dataset Shape & Basic Info
print (df.shape)
df.columns.tolist()
```

```python
df.info()
```

## Feature Group in Dataset
- **Demographics**
    - SEX,EDUCATION,MARRIAGE,AGE

- **Credit Information**
    - LIMIT_BAL - Credit Limit Amount

- **Payment History**
    - ( PAY_0 - PAY_6 ) - Past repayment status(delay months)

- **Bill Amount**
    - ( BILL_AMT1 - BILL_AMT6 ) - Monthly Bill Amount

- **Payment Amounts**
    - ( PAY_AMT1 to PAY_AMT6 ) - Previous Payment Amounts

- **Target Variable**
  - `default payment next month`
  - 1 → Default
  - 0 → No default

```python
# Finding Missing Value
missing_values = df.isnull().sum()
missing_values[missing_values>0]
print("Total Missing Values:", df.isnull().sum().sum())
```

```python
target_count=df['default payment next month'].value_counts()
target_count
```

```python
# Bar Plot
sns.countplot(x="default payment next month",data=df)
plt.title("Default Vs Non Default Distribution")
plt.xlabel("Default (1: Yes, 2: No)")
plt.ylabel("Count")
plt.show
```

```python
# Default Rate
default_rate=target_count[0]/len(df)
print(f"Default Rate: {default_rate:.2%}")
```

```python
df.describe()
```

```python
nums_col=['LIMIT_BAL','AGE']
for col in nums_col:
    sns.histplot(df[col],bins=30,kde=True)
    plt.title(f"Distribution of {col}")
    plt.show
```

```python
# Correlation Heatmap
plt.figure(figsize=(12,8))
corr_matrix=df.corr()
sns.heatmap(corr_matrix,cmap="coolwarm",center=0)
plt.title("Feature Correlation Heatmap")
plt.show()
```

#### Default Rate Based on Demographic

```python
# Default Rate Based on Sex
sns.barplot(x="SEX",y="default payment next month",data=df)
plt.title("Default Rate by Gender")
plt.show()
```

```python
# Default Rate Based on Education
sns.barplot(x="EDUCATION",y="default payment next month",data=df)
plt.title("Default Rate by Education")
plt.show()
```

```python
# Default Rate Based on Age
sns.histplot(data=df,x="AGE",bins=30, kde=True)
plt.title("Default Rate by Age")
plt.show()
```

```python
# Default Rate Based on Marriage
sns.barplot(x="MARRIAGE",y="default payment next month",data=df)
plt.title("Default Rate by Marriage")
plt.show()
```

```python
# Default Rate Based on Credit Limit
sns.barplot(x="default payment next month",y="LIMIT_BAL",data=df)
plt.title("Default Rate by Credit Limit")
plt.show()
```

## KEY INSIGHTS FOR EDA:
- Dataset Contains **0 Missing Value**, making preprocessing simpler.
- Default rate is approximately **77.88%%**, indicating **class imbalance**.  
- People with **Lower Credit Limit** has **Higher Default Rate**.
